```python
urbanization_data.rename(columns={'Iso3code': 'ISO3'}, inplace=True)
urbanization_data.rename(columns={'Area name': 'Country'}, inplace=True)

urbanization_data['Country'] = urbanization_data['Country'].str.upper()
air_pollution_data['Country'] = air_pollution_data['Country'].str.upper()


air_pollution_data.rename(columns={'Entity': 'Country'}, inplace=True)
air_pollution_data.rename(columns={'Code': 'ISO3'}, inplace=True)


who_air_quality_data['Country'] = who_air_quality_data['Country'].str.upper()

cities_air_water_data.rename(columns={' "Country"': 'Country'}, inplace=True)
cities_air_water_data.rename(columns={' "Region"': 'Region'}, inplace=True)
cities_air_water_data.rename(columns={' "AirQuality"': 'AirQuality'}, inplace=True)
cities_air_water_data.rename(columns={' "WaterPollution"': 'WaterPollution'}, inplace=True)

# Strip leading/trailing whitespace and replace 'United States of America' with 'United States'
cities_air_water_data['Country'] = cities_air_water_data['Country'].str.strip().replace('UNITED STATES OF AMERICA', 'UNITED STATES')


cities_air_water_data['Country'] = cities_air_water_data['Country'].str.replace('"', '')
cities_air_water_data['Country'] = cities_air_water_data['Country'].str.upper()
cities_air_water_data['Region'] = cities_air_water_data['Region'].str.replace('"', '')
cities_air_water_data['Region'] = cities_air_water_data['Region'].str.upper()

who_air_quality_data.rename(columns={'WHO Country Name': 'Country'}, inplace=True)
```

- as we can see above dataset has been modified to fit into the EDA process.

```python
def categorize_health_risk(pm2_5, pm10, no2):
    thresholds = {
        'pm2_5': {'low': 10, 'moderate': 35, 'high': 55, 'very_high': 150},
        'pm10': {'low': 20, 'moderate': 50, 'high': 80, 'very_high': 180},
        'no2': {'low': 25, 'moderate': 50, 'high': 100, 'very_high': 200},
    }
    
    risk_level_pm2_5 = next((risk for risk, value in thresholds['pm2_5'].items() if pm2_5 <= value), 'very_high')
    risk_level_pm10 = next((risk for risk, value in thresholds['pm10'].items() if pm10 <= value), 'very_high')
    risk_level_no2 = next((risk for risk, value in thresholds['no2'].items() if no2 <= value), 'very_high')
    
    combined_risk = max(risk_level_pm2_5, risk_level_pm10, risk_level_no2, key=lambda x: list(thresholds['pm2_5']).index(x))
    
    return combined_risk

urban_who['Health Risk'] = urban_who.apply(
    lambda row: categorize_health_risk(row['PM2.5 (μg/m3)'], row['PM10 (μg/m3)'], row['NO2 (μg/m3)']),
    axis=1
)
```

- the above code is for categorizing certain column for comparision degree as Health Risk.

### 5.7. Health Risk Distribution in Urban Areas

```python
iso_health_risks = urban_who.groupby(['ISO3_x', 'Residence', 'Health Risk']).size().reset_index(name='Counts')
iso_code = 'IND'
residence_type = 'Urban' 

iso_residence_health_risks = iso_health_risks[(iso_health_risks['ISO3_x'] == iso_code) & (iso_health_risks['Residence'] == residence_type)]

plt.figure(figsize=(8, 8))
plt.pie(iso_residence_health_risks['Counts'], labels=iso_residence_health_risks['Health Risk'], autopct='%1.1f%%', startangle=140)
plt.title(f'Health Risk Distribution for {iso_code} - {residence_type}')
plt.axis('equal') 
plt.show()
```
